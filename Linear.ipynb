```python
import matplotlib.pyplot as plt
import numpy as np

# Set up the figure and axis
fig, ax = plt.subplots()

# Draw the Pikachu face
circle = plt.Circle((0, 0), 1, color='gold')  # Pikachu's face
ax.add_artist(circle)

# Draw the eyes
left_eye = plt.Circle((-0.4, 0.3), 0.15, color='black')  # Left eye
ax.add_artist(left_eye)
right_eye = plt.Circle((0.4, 0.3), 0.15, color='black')  # Right eye
ax.add_artist(right_eye)

# Draw the nose
nose = plt.Polygon(np.array([[-0.1, 0.1], [0.1, 0.1], [0, -0.1]]), color='black')
ax.add_artist(nose)

# Draw the mouth
mouth = plt.Polygon(np.array([[-0.4, -0.2], [0.4, -0.2], [0.3, -0.4], [-0.3, -0.4]]), color='black')
ax.add_artist(mouth)

# Set the aspect ratio and limits
ax.set_aspect('equal')
ax.set_xlim(-1.2, 1.2)
ax.set_ylim(-1.2, 1.2)

# Hide the axis
ax.axis('off')

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create two 2D vectors
v1 = np.array([2, 3])
v2 = np.array([-1, 4])

# Plot each vector
plt.figure(figsize=(6, 6))
plt.plot([0, v1[0]], [0, v1[1]], color='r', label='v1')
plt.plot([0, v2[0]], [0, v2[1]], color='b', label='v2')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vectors')
plt.legend()
plt.grid(True)
plt.show()

# Perform arithmetic operations on the vectors
v_sum = v1 + v2
v_diff = v1 - v2
v_scalar = 4 * v1 + v2 / 2

# Plot v1+v2, v1-v2, v1*4+v2/2
plt.figure(figsize=(9, 6))

# v1+v2
plt.subplot(131)
plt.plot([0, v1[0]], [0, v1[1]], color='r', label='v1')
plt.plot([0, v2[0]], [0, v2[1]], color='b', label='v2')
plt.plot([0, v_sum[0]], [0, v_sum[1]], color='g', linestyle='--', label='v1+v2')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('v1 + v2')
plt.legend()
plt.grid(True)

# v1-v2
plt.subplot(132)
plt.plot([0, v1[0]], [0, v1[1]], color='r', label='v1')
plt.plot([0, -v2[0]], [0, -v2[1]], color='b', label='-v2')
plt.plot([0, v_diff[0]], [0, v_diff[1]], color='g', linestyle='--', label='v1-v2')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('v1 - v2')
plt.legend()
plt.grid(True)

# v1*4+v2/2
plt.subplot(133)
plt.plot([0, 4 * v1[0]], [0, 4 * v1[1]], color='r', label='4*v1')
plt.plot([0, v2[0] / 2], [0, v2[1] / 2], color='b', label='v2/2')
plt.plot([0, v_scalar[0]], [0, v_scalar[1]], color='g', linestyle='--', label='4*v1 + v2/2')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('4*v1 + v2/2')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create three 2D vectors
v1 = np.array([3, 0])  # Orthogonal vector 1
v2 = np.array([0, 2])  # Orthogonal vector 2
v3 = np.array([2, 2])  # Non-orthogonal vector

# Plot the three vectors
plt.figure(figsize=(6, 6))
plt.plot([0, v1[0]], [0, v1[1]], color='r', label='v1')
plt.plot([0, v2[0]], [0, v2[1]], color='b', label='v2')
plt.plot([0, v3[0]], [0, v3[1]], color='g', label='v3')
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vectors')
plt.legend()
plt.gca().set_aspect('equal')
plt.grid(True)
plt.show()
```
